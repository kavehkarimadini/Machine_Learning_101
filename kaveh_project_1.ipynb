## 5)get the input from client and generate a fibonacci sequence till that input

 

```python
client = input("how much does the end means to you? =>")
```

```python
# Program to display the Fibonacci sequence up to n-th term
# first two terms
n1, n2 = 0, 1
count = 0
client = int(client)
#--------------------------------#
# check if the number of terms is valid
if (client <= 0):
   print("Please enter a positive integer")
# if there is only one term, return n1
elif client == 1:
   print("Fibonacci sequence upto",client,":")
   print(n1)
# generate fibonacci sequence
else:
   print("Fibonacci sequence:")
   while count < client:
       print(n1)
       nth = n1 + n2
       # update values
       n1 = n2
       n2 = nth
       count += 1
```
