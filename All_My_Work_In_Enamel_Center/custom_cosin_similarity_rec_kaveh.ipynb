```python
#import sklearn library
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
df_rating = pd.read_csv('enamel_clean_dataset.csv')
```

```python
df_product = pd.read_csv('product_clean_dataset.csv')
```

```python
df_rating_two = df_rating.groupby(['product._id','user._id'])['product._id'].count().unstack().reset_index().fillna(0).set_index('product._id')
```

```python
df_rating_two
```

```python
def rating_df(df):
    for i in range(len(df.index)):
        row = df.iloc[i]
        sum_row = row.sum()
        if sum_row == 0:
            continue
        for j in range(len(df.columns)):
            if df.iloc[i,j] == 0:
                continue
            else:
                df.iloc[i,j] = df.iloc[i,j]/sum_row
        return df
```

```python
df_rating_three = rating_df(df_rating_two)
```

```python
df_product = df_product.drop(columns=['Unnamed: 0']).drop_duplicates()
```

```python
df_product.isna().sum()
```

```python
def dummies(x,df):
    temp = pd.get_dummies(df[x], drop_first = True)
    df = pd.concat([df, temp], axis = 1)
    df.drop([x], axis = 1, inplace = True)
    return df
```

```python
df_encoded_product = dummies('product.texture',df_product)
```

```python
df_encoded_product = dummies('product.usage_of_product',df_encoded_product)
```

```python
df_encoded_product = dummies('product.category',df_encoded_product)
```

```python
need = [ 'product.price', 'product.weight',
       'product.height', 'product.width', 'product.length']
scaler = MinMaxScaler()
df_encoded_product[need]=scaler.fit_transform(df_encoded_product[need])
```

```python
df_encoded_one = df_encoded_product.set_index('product._id',drop = True)
```

```python
df_encoded_one.columns
```

```python
df_encoded_one.drop(columns=['product.name'],inplace=True)
```

```python
df_clean_semi_final = pd.merge(df_encoded_one, df_rating_three, left_index=True, right_index=True)
```

```python
df_clean_semi_final=df_clean_semi_final.drop_duplicates(subset='product._id').set_index('product._id',drop = True)
```

```python
def item_similarity(item1,item2,dataset):
    item1_ratings = [dataset.loc[item1]]
    item2_ratings = [dataset.loc[item2]]
    cs = cosine_similarity(item1_ratings,item2_ratings)
    return cs
```

```python
dic = {}
for row1 in df_clean_semi_final.index:
    item1_ratings = [df_clean_semi_final.loc[row1]]
    my_list=[]
    for row2 in df_clean_semi_final.index:
        item2_ratings = [df_clean_semi_final.loc[row2]]
        cs = cosine_similarity(item1_ratings,item2_ratings)
        my_list.append(cs)
    dic[row1] = my_list    
```

```python
df_final_item_item = pd.DataFrame(dic,index = df_clean_semi_final.index )
```
