```python
df_rating = pd.read_csv('enamel_clean_dataset.csv')
```

```python
df_two = df_rating.groupby(['product._id','user._id'])['product._id'].count().unstack().reset_index().fillna(0).set_index('product._id')
```

```python
def rating_df(df):
    for i in range(len(df.index)):
        row = df.iloc[i]
        sum_row = row.sum()
        if sum_row == 0:
            continue
        for j in range(len(df.columns)):
            if df.iloc[i,j] == 0:
                continue
            else:
                df.iloc[i,j] = df.iloc[i,j]/sum_row
        return df
```

```python
df_three = rating_df(df_two)
```

```python
df_four = df_three.reset_index
```

```python
df_three.index[0]
```

```python
item_dup = []
user = []
rating = []
for row in range(len(df_three.index)):
    for column in range(len(df_three.columns)):
        item_dup.append(df_three.index[row])
        user.append(df_three.columns[column])
        rating.append(df_three.iloc[row,column])
```

```python
len(item_dup) == len(user)
```

```python
len(user)==len(rating)
```

```python
df_clean_rating = pd.DataFrame({
    'user_id': user,
    'item_id': item_dup,
    'rating': rating
})
```
