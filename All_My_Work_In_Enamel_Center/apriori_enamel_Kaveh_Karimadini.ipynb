```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_excel('data.xlsx',sheet_name='Sheet1')
```

```python
df
```

```python
basket_France = (df.groupby(['Name','kala'])['number'].sum().unstack().reset_index().fillna(0).set_index('Name'))
```

```python
basket_France
```

```python
# Defining the hot encoding function to make the data suitable 
# for the concerned libraries
def hot_encode(x):
  if x<= 0:
    return 0  
  else:
    return 1
basket_encoded = basket_France.applymap(hot_encode)
basket_France = basket_encoded
```

```python
q = apriori(basket_France, min_support=0.001, use_colnames=True, max_len=3)
#q['length'] = q['itemsets'].apply(len)
q.iloc[:100]
```

```python
q
```

```python
# Collecting the inferred rules in a dataframe
rules = association_rules(q,  metric="confidence", min_threshold=0.7)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
rules.head(20)
```
